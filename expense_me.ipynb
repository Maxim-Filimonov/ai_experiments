```python
import gspread
import os
import openai
from oauth2client.service_account import ServiceAccountCredentials

def magic_expense(user_prompt):

  # Define scope
  scope = ['https://spreadsheets.google.com/feeds',
          'https://www.googleapis.com/auth/drive']

  # Add credentials to the account
  creds = ServiceAccountCredentials.from_json_keyfile_name(
      '/Users/maxim/Downloads/micro-environs-208817-b161a093d327.json', scope)

  # Authorize the clientsheet
  client = gspread.authorize(creds)

  # Get the instance of the Spreadsheet
  sheet = client.open('My Budget')

  # Get Income sheet
  income_sheet = sheet.get_worksheet(0)

  # Get Expenses sheet
  expenses_sheet = sheet.get_worksheet(1)

  openai.api_key = os.environ["OPENAI_API_KEY"]

  response = openai.Completion.create(
      model="text-davinci-003",
      prompt=f"Convert this text to a programmatic command:\n\nExample: Record that I have spent 90 AUD on dinner\nexpenses_sheet.append_row([90, \"AUD\", \"Eating\", \"26-02-2023\"])\nExample: Went to bar for 50$ last night\nexpenses_sheet.append_row([50, \"USD\", \"Eating\", \"25-02-2023\"])\nExample: Got my salary 5000 AUD today!\nexpenses_sheet.append_row([5000, \"AUD\", \"Salary\", \"26-02-2023\"])\nExample: Got my cheque for 7000 AUD last night!\nincome_sheet.append_row([5000, \"AUD\", \"Salary\", \"25-02-2023\"])\nExample: Received money for my hardwork last week on Monday 3500 USD\nincome_sheet.append_row([3500, \"USD\", \"Salary\", \"13-02-2023\"])\n\n{user_prompt}",
      temperature=0,
      max_tokens=100,
      top_p=1,
      frequency_penalty=0.2,
      presence_penalty=0
  )

  # if there are choices pic the first one
  if response["choices"]:
      first_choice = response["choices"][0]["text"]
      firsts_choice = first_choice.replace("\n", "")
      print("running", first_choice)
      # evaluate the choice as python code
      eval(first_choice)
```

```python
# user prompt
user_prompts = [
    "Record that I have spent 90 AUD on dinner",
    "Went to bar for 50$ last night",
    "Woohoo got salary for 10000 EUR three days ago",
    "Smashed the bar for 100$ last night",
    "Приехала зарплата 10000 RUB вчера",
]
# for each user prompt
for user_prompt in user_prompts:
    # run the magic expense function
    magic_expense(user_prompt)
```
